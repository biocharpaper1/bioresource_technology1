```python
#Assuming H0: The difference in Adsorption Capacity means wrt two different Biomass is zero
#Assuming H1: The difference in Adsorption Capacity means wrt two different Biomass is not zero
#creating two dtaframes from the existing dataframe
import pandas as pd

# Assuming your existing DataFrame is named 'df'
# and the column containing the class labels is named 'class_label'

# Replace 'desired_class_label' with the actual class label you want to filter for
desired_class_label1 = 'CSW'
desired_class_label2 = 'SP'
# Filter the DataFrame to include only rows with the desired class label
ds5_filt_CSW = ds5_filt[ds5_filt['Biomass'] == desired_class_label1]
print(ds5_filt_CSW)
ds5_filt_SP = ds5_filt[ds5_filt['Biomass'] == desired_class_label2]
print(ds5_filt_SP)
```

```python
#calculating mean of two groups
print(ds5_filt_CSW['Adsorption Capacity'].mean())
print(ds5_filt_SP['Adsorption Capacity'].mean())
```

```python
#independent T-test
from scipy.stats import ttest_ind
stat, p = ttest_ind(ds5_filt_CSW['Adsorption Capacity'], ds5_filt_SP['Adsorption Capacity'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
if p>0.05:
  print('Null Hypothesis accepted')
  print('The difference in Adsorption Capacity means wrt two different Biomass is zero')
else:
  print('Alternate Hypothesis accepted')
  print('The difference in Adsorption Capacity means wrt two different Biomass is not zero')
```

```python
#creating a new dtaframe from the existing dataframe
desired_class_label1 = 'CSW'
desired_class_label2 = 'SP'
desired_class_label3 = 'M'
ds5_filt_CSW = ds5_filt[ds5_filt['Biomass'] == desired_class_label1]
print(ds5_filt_CSW)
ds5_filt_SP = ds5_filt[ds5_filt['Biomass'] == desired_class_label2]
print(ds5_filt_SP)
ds5_filt_M = ds5_filt[ds5_filt['Biomass'] == desired_class_label3]
print(ds5_filt_M)

#calculating mean of the group
print(ds5_filt_CSW['Adsorption Capacity'].mean())
print(ds5_filt_SP['Adsorption Capacity'].mean())
print(ds5_filt_M['Adsorption Capacity'].mean())
```

```python
#ANOVA
#Assuming H0: The difference in Adsorption Capacity means wrt three different Biomass is zero
#Assuming H1: The difference in Adsorption Capacity means wrt three different Biomass is not zero

from scipy.stats import f_oneway
stat, p = f_oneway(ds5_filt_CSW['Adsorption Capacity'], ds5_filt_SP['Adsorption Capacity'], ds5_filt_M['Adsorption Capacity'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
if p>0.05:
  print('Null Hypothesis accepted')
  print('Same distribution of score')
else:
  print('Alternate Hypothesis accepted')
  print('Different distribution of score')
```

```python
#Mann-Whitney U Test
#Assuming H0: The difference in Adsorption Capacity means wrt two different Biomass is zero
#Assuming H1: The difference in Adsorption Capacity means wrt two different Biomass is not zero
from scipy.stats import mannwhitneyu
stat, p = mannwhitneyu(ds5_filt_CSW['Adsorption Capacity'], ds5_filt_M['Adsorption Capacity'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
if p>0.05:
  print('Null Hypothesis accepted')
  print('Same distribution of score')
else:
  print('Alternate Hypothesis accepted')
  print('Different distribution of score')
```
